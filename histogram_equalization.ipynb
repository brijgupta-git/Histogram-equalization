# implementation of histogram equalization

```python
from PIL import Image 
import numpy as np
from matplotlib import pyplot as plt

def hist_table(im):
        arr=np.array(im)
        farr=arr.flat      
        intensity,frequency=np.unique(farr,return_counts=True)     
        table=np.array((intensity,frequency)).T  
        N=np.sum(table[0:,1:])
        table=np.column_stack((table,np.divide(table[0:,1:],N)))
        table=np.column_stack((table,np.cumsum(table[0:,2:])))
        table=np.column_stack( (table,np.multiply(table[0:,3:],255)))
        table=np.column_stack( (table,np.round(table[0:,4:],0)))
        return table

def hist_eq(table,pv):
        r=np.where(table[0:,0]==pv)
        try:
            return int(table[r[0][0],-1])
        except:
            return pv

def histogram_equalization(im):
    table=hist_table(im)
    h_im=im.point(lambda x:  hist_eq(table,x))
    return h_im
    
```

```python
im1=Image.open('images/1.tif').convert("L")
im1
```

**input image 1**

```python
arr1=np.array(im1)
plt.hist(arr1.flat)
```

**histgram graph for input image 1**

```python
h_im1=histogram_equalization(im1)
h_im1
```

**output image1**

```python
h_arr1=np.array(h_im1)
plt.hist(h_arr1.flat)
```

**histgram graph for output image 1**
